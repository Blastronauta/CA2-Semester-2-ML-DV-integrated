Feature engineer for Recommender Systems
---
In this section we prepare the data required for the recommender system models. 
We use the reduced dataset "orders_products_small" consisting of 5,000 users to ensure efficient computation. 

We perform three key tasks:

1. Creating item-level features for content-based filtering  
2. Building the useritem interaction matrix  
3. Preparing data structures for useruser and itemitem collaborative filtering

First we extract unique products fromreduced dataset for TF-IDF  matrix creation:

```python
products_small = orders_products_small[['product_id', 'product_name', 'department_name']].drop_duplicates()

print("Unique products in reduced dataset:", products_small.shape[0])
products_small.head()
```

Matrix construction:

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(products_small['product_name'])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

Now we have the matrix ready for content base filtering, now lets take the orders to get purchase count with user and product variables:

```python
user_product_counts = (
    orders_products_small
    .groupby(['user_id', 'product_id'])['order_id']
    .count()
    .reset_index()
)

user_product_counts.rename(columns={'order_id': 'purchase_count'}, inplace=True)
user_product_counts.head()
```

Now we pivot to create User-Item Matrix:

```python
user_item_matrix = user_product_counts.pivot(
    index='user_id',
    columns='product_id',
    values='purchase_count'
)

user_item_matrix = user_item_matrix.fillna(0)

user_item_matrix.shape
```

now for performance we are going to use sparse:

```python
user_item_sparse = csr_matrix(user_item_matrix.values)
user_item_sparse
```

Content-Based Filtering
---
This model recommends products that are textually similar, using only product names.
We apply TF-IDF (Term FrequencyInverse Document Frequency) to transform product names into numerical vectors and compute cosine similarity to identify similar items.

Content-based filtering does not require any user purchase history, making it ideal for cold-start situations and browsing assistance.

```python
# Reset index to match TF-IDF matrix
products_small = products_small.reset_index(drop=True)

# product_id -> tfidf row index
product_to_index = pd.Series(products_small.index, index=products_small['product_id'])

# tfidf row index -> product_id
index_to_product = pd.Series(products_small['product_id'], index=products_small.index)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
def _recommend_similar_products_by_id(product_id, n=5):
    """
    Internal helper: recommend top-n similar products given a product_id.
    """
    if product_id not in product_to_index:
        print("Product ID not found in sample dataset.")
        return None
    
    idx = product_to_index[product_id]
    
    # Compute similarity scores
    scores = list(enumerate(cosine_sim[idx]))
    
    # Sort by similarity (skip itself)
    scores = sorted(scores, key=lambda x: x[1], reverse=True)[1:n+1]
    
    indices = [i[0] for i in scores]
    similar_ids = index_to_product[indices].values
    
    recs = products_small[products_small['product_id'].isin(similar_ids)].copy()
    return recs[['product_id', 'product_name']]
```

```python
def recommend_similar_by_keyword(keyword, n=5):
    """
    Main content-based recommender.

    1. Searches for products whose name contains the keyword.
    2. Picks the first match (you could change this rule).
    3. Uses TF-IDF cosine similarity to recommend n similar products.
    """
    if not keyword or not isinstance(keyword, str):
        print("Please provide a non-empty keyword (e.g., 'milk', 'salted butter').")
        return None
    
    # Find products whose name contains the keyword (case-insensitive)
    matches = products_small[products_small['product_name'].str.contains(keyword, case=False, na=False)]
    
    if matches.empty:
        print(f"No products found containing '{keyword}'. Try a different word.")
        return None
    
    # Here we pick the first match; you could improve this (e.g., choose most popular)
    base_product = matches.iloc[0]
    base_id = base_product['product_id']
    
    print("Base product used for recommendations:")
    print(f"- ID:   {base_id}")
    print(f"- Name: {base_product['product_name']}\n")
    
    # Use the internal ID-based recommender
    recs = _recommend_similar_products_by_id(base_id, n=n)
    
    print("Recommended similar products:\n")
    return recs
```

Collaborative Filtering
---
In this section we implement both **ItemItem Collaborative Filtering** and **UserUser Collaborative Filtering**, using the reduced dataset of 5,000 users.

Collaborative filtering relies on purchase patterns: users who buy similar products are considered similar, and products bought by similar users are likely to be recommended.

We use the useritem interaction matrix created earlier, where:
- Rows = users  
- Columns = products  
- Values = purchase counts

Item-Item Collaborative Filtering
---
This model finds products that are purchased by similar groups of users

```python
# Transpose: products x users
item_user_matrix = user_item_matrix.T  

# Compute cosine similarity between items
item_item_sim = cosine_similarity(item_user_matrix)

item_item_sim.shape
```

```python
# Map product_id -> matrix row
item_index = pd.Series(range(len(item_user_matrix.index)), index=item_user_matrix.index)

# Map matrix row -> product_id
index_to_item = pd.Series(item_user_matrix.index, index=range(len(item_user_matrix.index)))
```

This next code is an internal function using product ID to identified the product, next we will get a more user friendly interface

```python
def _recommend_item_item_by_id(product_id, n=5):
    """
    Internal helper: recommend top-n similar items given a product_id,
    using itemitem collaborative filtering.
    """
    if product_id not in item_index:
        print("Product not found in the reduced dataset.")
        return None
    
    idx = item_index[product_id]
    
    # Similarity scores for this item vs all others
    sim_scores = list(enumerate(item_item_sim[idx]))
    
    # Sort by similarity, skip itself
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n+1]
    
    similar_indices = [i[0] for i in sim_scores]
    similar_ids = index_to_item[similar_indices].values
    
    # Look up names
    recs = products_small[products_small['product_id'].isin(similar_ids)].copy()
    return recs[['product_id', 'product_name']]
```

```python
def recommend_item_item_by_name(keyword, n=5):
    """
    Public function: ItemItem CF using a product name keyword.
    
    1. Search for products whose names contain the keyword.
    2. Pick the first match as the base product.
    3. Use itemitem collaborative filtering to recommend similar items.
    """
    if not keyword or not isinstance(keyword, str):
        print("Please provide a non-empty keyword (e.g., 'Salted Butter', 'milk').")
        return None
    
    # Find matching products by name (case-insensitive)
    matches = products_small[products_small['product_name'].str.contains(keyword, case=False, na=False)]
    
    if matches.empty:
        print(f"No products found containing '{keyword}'. Try another term.")
        return None
    
    # Choose the first match (you could later refine this to pick the most popular)
    base_product = matches.iloc[0]
    base_id = base_product['product_id']
    
    print("Base product used for ItemItem CF:")
    print(f"- ID:   {base_id}")
    print(f"- Name: {base_product['product_name']}\n")
    
    recs = _recommend_item_item_by_id(base_id, n=n)
    print("Recommended similar products (ItemItem CF):\n")
    return recs
```

```python
recommend_item_item_by_name("Salted Butter", n=6)
```

ItemItem collaborative filtering identifies products that tend to be purchased by the same users.

To provide a user-friendly interface, we allow the user (or dashboard) to input a **product name keyword** (e.g., "Salted Butter").  
The system finds a matching product and then uses purchase patterns to recommend other items that similar customers bought.

This approach captures behavioural relationships between items that may not be obvious from the product name alone.

User-User Collaborative Filtering
---
Here we use a similarity of some user's products plus recommendations, using user_id is normal and expected: we dont have real names, and the IDs are anonymous.

```python
# Cosine similarity between users on the user_item_matrix
user_user_sim = cosine_similarity(user_item_matrix)

user_user_sim.shape
```

```python
user_index = pd.Series(range(len(user_item_matrix.index)), index=user_item_matrix.index)
index_to_user = pd.Series(user_item_matrix.index, index=range(len(user_item_matrix.index)))
```

```python
def recommend_user_user(user_id, n_neighbors=5, n_recommendations=5):
    """
    Recommend items to a user based on similar users (UserUser CF).
    
    Steps:
    1. Find the most similar users.
    2. Aggregate the items they bought.
    3. Remove items the target user already purchased.
    4. Recommend the top remaining products.
    """
    if user_id not in user_index:
        print("User not found in reduced dataset.")
        return None
    
    # Get index in similarity matrix
    idx = user_index[user_id]
    
    # Similarity scores to all other users
    sim_scores = list(enumerate(user_user_sim[idx]))
    
    # Sort by similarity and take top neighbors (excluding user itself)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:n_neighbors+1]
    
    similar_user_indices = [i[0] for i in sim_scores]
    similar_users = index_to_user[similar_user_indices].values
    
    # Items bought by similar users
    similar_user_data = user_item_matrix.loc[similar_users]
    
    # Sum up purchase counts across similar users
    item_scores = similar_user_data.sum(axis=0)
    
    # Remove items already bought by the target user
    user_items = user_item_matrix.loc[user_id]
    item_scores = item_scores[user_items == 0]
    
    # Top-N recommended items
    top_items = item_scores.sort_values(ascending=False).head(n_recommendations)
    
    # Lookup readable product names
    recommendations = products_small[products_small['product_id'].isin(top_items.index)][['product_id', 'product_name']]
    
    # Also show some of the user's existing items for context
    user_purchased_ids = user_items[user_items > 0].index
    user_purchased_sample = products_small[products_small['product_id'].isin(user_purchased_ids)].head(10)
    
    return user_purchased_sample, recommendations
```

```python
# Pick a random user from the reduced dataset
example_user = user_item_matrix.sample(1).index[0]
print("Selected user_id:", example_user)

user_items, user_recs = recommend_user_user(example_user)

print("\nSample of items already purchased by this user:\n")
display(user_items)

print("\nRecommended new items for this user (UserUser CF):\n")
display(user_recs)
```

Market Basket Analysis
---
In this section we apply Market Basket Analysis on the Instacart data:
- We treat each order as a **transaction** (basket of products).
- We use the **Apriori algorithm** to find frequent itemsets.
- We generate **association rules** with support, confidence and lift.

To keep the computation efficient, we work with the reduced dataset (`orders_products_small`, 5,000 users) and focus on the most frequently purchased products.

First to Apriori we are going to use top 20 products and 5000 orders max with only top products:

```python
TOP_N_PRODUCTS = 20      # keep MBA light and efficient
MAX_ORDERS = 5000        # number of orders to sample

# 1) Select top-N most frequent products in the reduced dataset
top_products = (
    orders_products_small['product_name']
    .value_counts()
    .head(TOP_N_PRODUCTS)
)

print("Top products used in MBA:")
display(top_products)

# 2) Keep only rows containing these top products
mba_data = orders_products_small[
    orders_products_small['product_name'].isin(top_products.index)
].copy()

print("Rows after filtering to top products:", mba_data.shape)

# 3) Sample a manageable number of orders
unique_orders = mba_data['order_id'].unique()
n_orders = min(MAX_ORDERS, len(unique_orders))

print("Unique orders available after filtering:", len(unique_orders))
print("Sampling this many orders:", n_orders)

np.random.seed(42)
sampled_orders = np.random.choice(unique_orders, size=n_orders, replace=False)

mba_sample = mba_data[mba_data['order_id'].isin(sampled_orders)].copy()

print("Final MBA sample shape:", mba_sample.shape)
print("Unique orders in sample:", mba_sample['order_id'].nunique())
print("Unique products in sample:", mba_sample['product_name'].nunique())
```

Then we will have to change only for this function the data type of product id as object an then we create the basket as a boolean:

```python
# Use product_id as the item identifier
mba_sample = mba_sample.copy()
mba_sample['product_id'] = mba_sample['product_id'].astype(str)

basket = (
    mba_sample
    .groupby(['order_id', 'product_id'], observed=True)['product_id']
    .count()
    .unstack(fill_value=0)
)
basket_bool = basket > 0

print("Basket shape:", basket_bool.shape)
basket_bool.head()
```

Apriori Algorithm
---
Now we are going to apply the algorithm:

```python
MIN_SUPPORT=0.03
frequent_itemsets = apriori(
    basket_bool,
    min_support=MIN_SUPPORT,  
    use_colnames=True
)
print("Itemsets:", frequent_itemsets.shape[0])
```

Using the Apriori algorithm, we identified frequent itemsets among the most popular products in our reduced dataset. 
With an initial minimum support of **0.03**, the model produced around **22 frequent itemsets**, indicating a relatively high number of product combinations that co-occur in at least 3% of sampled orders.

- A **lower support** threshold includes more rare combinations, producing many itemsets.  
- A **higher support** threshold filters out weak or less relevant associations, leaving a smaller set of meaningful combinations.
- Instacart baskets tend to be relatively small, so only a few product pairs appear frequently across many orders.

These 22 itemsets represent the **core co purchasing patterns** in our sample and will be used to generate association rules in the next step.

Asociation Rules
---
First we are going to configure a threshold of 1 for Lift:

```python
rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=0.7
)

print("Number of rules generated:", rules.shape[0])
rules.head()
```

Since we didnt get any rules with 1 or 0.7 forLift metric now Im going to try with confidence >=0.3

Now since we got 4 rules we are going to use product names so we are going to replace those product ids:
